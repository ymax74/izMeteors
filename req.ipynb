запрос к БД на сайте izmeteors.puldb.ru

```python
import requests
import pandas as pd
from io import StringIO

from astropy.time import Time
```

```python
url = 'https://izmeteors.puldb.ru/req.php?q='
# r = requests.get(f'{url}DESCRIBE izMeteors_meteor_data_tbl;').text
# print(r)
```

```python
mjd1 = Time('2025-10-07T15:00:00.0',format='isot').copy(format='mjd').value
mjd2 = Time('2025-10-08T03:00:00.0',format='isot').copy(format='mjd').value
```

```python
r = requests.get(f'{url}SELECT * FROM izMeteors_meteor_data_tbl WHERE mjd_beg>{mjd1} AND mjd_beg<{mjd2} AND obs_id LIKE "Astro%" AND cam_id LIKE "15";').text
df = pd.read_csv(StringIO(r))
df
```

```python
for index, row in df.iterrows():
    print(f'http://sfa.puldb.ru:9020/{row['dir_name']}/m{row['base_id']}.mp4')
```
